### Columns:
customer_ID - A unique identifier for the customer

shopping_pt - Unique identifier for the shopping point of a given customer

record_type - 0=shopping point, 1=purchase point

day - Day of the week (0-6, 0=Monday)

time - Time of day (HH:MM)

state - State where shopping point occurred

location - Location ID where shopping point occurred

group_size - How many people will be covered under the policy (1, 2, 3 or 4)

homeowner - Whether the customer owns a home or not (0=no, 1=yes)

car_age - Age of the customerâ€™s car

car_value - How valuable was the customerâ€™s car when new

risk_factor - An ordinal assessment of how risky the customer is (1, 2, 3, 4)

age_oldest - Age of the oldest person in customer's group

age_youngest - Age of the youngest person in customerâ€™s group

married_couple - Does the customer group contain a married couple (0=no, 1=yes)

C_previous - What the customer formerly had or currently has for product option C (0=nothing, 1, 2, 3,4)

duration_previous -  how long (in years) the customer was covered by their previous issuer

A,B,C,D,E,F,G - the coverage options

cost - cost of the quoted coverage options

```python
data_h2o = pd.read_csv('Data/train.csv')
```

```python
cols = ['A', 'B', 'C','D', 'E', 'F', 'G']
```

```python
for i in cols:
    data_h2o[i] = data[i].astype(str)
```

```python
data_h2o["Policy"] = data_h2o[cols].apply(lambda x: ''.join(x), axis=1)
```

```python
data_h2o.Policy
```

```python
data_h2o["Policy"] = data_h2o["Policy"].astype(int)
```

```python
data_h2o.Policy
```

```python
data_h2o.head(20)
```

```python
data_h2o.drop(cols, axis=1)
```

```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
df= h2o.H2OFrame(data_h2o)
```

```python
splits = df.split_frame(ratios=[0.7, 0.15], seed=1)
```

```python
train = splits[0]
valid = splits[1]
test = splits[2]
```

```python
x = train.columns
y= 'record_type'
```

```python
x.remove(y)
```

```python
aml = H2OAutoML()
```

```python
aml.train(x = x, y = y, training_frame = train, leaderboard_frame= test, validation_frame=valid)
```

```python
lb = aml.leaderboard
```

```python
lb
```

```python
leader = aml.leader
```

```python
import matplotlib as plt
%matplotlib inline

leader.varimp_plot
```

```python
aml.leader.varimp_plot()
```

```python
leader
```

```python
data_test = pd.read_csv('Data/test_v2.csv')
```

```python
for i in cols:
    data_test[i] = data_test[i].astype(str)
```

```python
data_test["Policy"] = data_test[cols].apply(lambda x: ''.join(x), axis=1)
```

```python
data_test["Policy"] = data_test["Policy"].astype(int)
```

```python
data_test.drop(cols, axis=1)
```

```python
df_test = h2o.H2OFrame(data_test)
```

```python
h2o_preds = aml.predict(test_data = df_test)
```

```python
sub = df_test.cbind(h2o_preds)
sub = sub[['customer_ID', 'Policy', 'predict']]
sub.set_names(['customer_ID','Policy', 'record_type' ])
```

```python
import os
h2o.export_file(sub[['customer_ID','Policy']],os.getcwd()+'\\'+'initial_Submission1.csv')
```

```python
submit_df = pd.read_csv('initial_Submission1.csv')
```

```python
submit_df.head()
```

```python
selection = []
count = 0
while (count< submit_df.shape[0]-1):
    if not int(submit_df['customer_ID'][count]) == int(submit_df['customer_ID'][count+1]):
        selection.append(submit_df['Policy'][count])
    count+=1
```

```python
import numpy as np
cust_ID_list = submit_df['customer_ID'].unique()
np.count_nonzero(cust_ID_list)
```

```python
np.count_nonzero(selection)
```

```python
selection.append(submit_df['Policy'][count])
```

```python
submission = pd.DataFrame(
    {'customer_ID': cust_ID_list,
     'plan_select': selection
    })
```

```python
submission.to_csv("initial_submit.csv")
```
