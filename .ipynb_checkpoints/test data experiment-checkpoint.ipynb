```python
selection = []
count = 0
while (count< df.shape[0]-1):
    if not int(df['customer_ID'][count]) == int(df['customer_ID'][count+1]):
        a_sel = str(df['A'][count])
        b_sel = str(df['B'][count])
        c_sel = str(df['C'][count])
        d_sel = str(df['D'][count])
        e_sel = str(df['E'][count])
        f_sel = str(df['F'][count])
        g_sel = str(df['G'][count])
        choice = str(a_sel +b_sel +c_sel +d_sel +e_sel +f_sel +g_sel)
        selection.append(choice)
    count+=1
```

```python
a_sel = str(df['A'][count])
b_sel = str(df['B'][198855])
c_sel = str(df['C'][198855])
d_sel = str(df['D'][198855])
e_sel = str(df['E'][198855])
f_sel = str(df['F'][198855])
g_sel = str(df['G'][198855])
choice = a_sel +b_sel +c_sel +d_sel +e_sel +f_sel +g_sel
selection.append(choice)
```

```python
cust_ID_list = df['customer_ID'].unique()
np.count_nonzero(cust_ID_list)
```

```python
np.count_nonzero(selection)
```

```python
submission = pd.DataFrame(
    {'customer_ID': cust_ID_list,
     'plan_select': selection
    })
```

```python
submission.to_csv("initial_submit.csv")
```
